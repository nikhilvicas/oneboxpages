```python
cachedCoordinates = [0, 0, 0, 0]
```

```python
def boxpoints(h, w):
    point1x = round(random.uniform(0, w), 2) # np.random.randint(0, w)
    point1y = round(random.uniform(0, h), 2) # np.random.randint(0, h)
    deltax = 0 # width of the box
    deltay = 0 # height of the box
    deltax = round(random.uniform(50, w - point1x), 2) # np.random.randint(0, w - point1x)
    deltay = round(random.uniform(50, w - point1x), 2) # np.random.randint(0, h - point1y)
    if (point1x + deltax > w):
        point2x = w # max(w, point1x + deltax)
    else:
        point2x = point1x + deltax
    if (point1y + deltay > h):
        point3y = h
    else:
        point3y = point1y + deltay
    return point1x, point2x, point1y, point3y
```

```python
def makebox(num):
    h, w = 1344, 932
    data = np.zeros((h, w, 3), dtype=np.uint8)
    minX, maxX, minY, maxY = boxpoints(h, w)
    # print(minX, maxX, minY, maxY)
    data[int(minY):int(maxY), int(minX):int(maxX)] = [255, 255, 255] # white box in coordinates
    img = im.fromarray(data, 'RGB')
    filename = '/users/nikhilvicas/Desktop/personal/nlmatics/modeltesting/box/images/train/page' + str(num) + '.png'
    img.save(filename)
    # img.show()
    global cachedCoordinates 
    cachedCoordinates = [minX, maxX, minY, maxY]
```

```python
def boxscale():
    h, w = 1344, 932
    global cachedCoordinates
    # print('here')
    # print(cachedCoordinates)
    minX, maxX, minY, maxY = cachedCoordinates
    minXscaled = minX/w
    maxXscaled = maxX/w
    minYscaled = minY/h
    maxYscaled = maxY/h
    # print(minXscaled, minYscaled, maxXscaled, maxYscaled)
    return minX, maxX, minY, maxY, minXscaled, minYscaled
```

```python
def coordinates():
    h, w = 1344, 932
    minX, maxX, minY, maxY, minXscaled, minYscaled = boxscale()
    # its like that scene in inception when they realize they have to go four dreams deep to get saito
    width = (maxX - minX)/w
    height = (maxY - minY)/h
    # print(minXscaled, minYscaled, width, height)
    return minXscaled, minYscaled, width, height
```

```python
def addlabel():
    label = 0 # np.random.randint(0, high = 10)
    minXscaled, minYscaled, width, height = coordinates()
    # print(label, minXscaled, minYscaled, width, height)
    return label, minXscaled, minYscaled, width, height
```

```python
for i in range(200):
    filename = '/users/nikhilvicas/Desktop/personal/nlmatics/modeltesting/box/labels/train/label' + str(i) + '.txt'
    print(filename)
    makebox(i)
    label, minXscaled, minYscaled, width, height = addlabel()
    vals = [label, minXscaled, minYscaled, width, height]
    np.savetxt(filename, vals, fmt='%s', newline = " ")
```
